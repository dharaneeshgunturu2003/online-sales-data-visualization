Importing necessary libraries

```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Online Sales Data.csv')
```

```python
df.head(5)
```

Bar plot

```python
category_revenue = df.groupby('Product Category')['Total Revenue'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(category_revenue['Product Category'], category_revenue['Total Revenue'], color='skyblue')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Product Category')
plt.xticks(rotation=45)
plt.show()
```

Stacked bar plot

```python
# Aggregate the data by Region and Product Category
category_region = df.groupby(['Region', 'Product Category']).size().unstack(fill_value=0)

# Plot
category_region.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Region')
plt.ylabel('Number of Units Sold')
plt.title('Units Sold by Product Category and Region')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Heat Map

```python
category_region = df.groupby(['Region', 'Product Category']).size().unstack(fill_value=0)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(category_region, annot=True, fmt="d", cmap="YlGnBu")

plt.xlabel('Product Category')
plt.ylabel('Region')
plt.title('Heatmap of Units Sold by Product Category and Region')
plt.show()
```

Histogram

```python
payment_method_counts = df['Payment Method'].value_counts()
plt.figure(figsize=(10, 6))
payment_method_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Payment Method')
plt.ylabel('Frequency')
plt.title('Distribution of Payment Methods')
plt.xticks(rotation=45)
plt.show()
```

Box Plot

```python
plt.figure(figsize=(12, 6))
df.boxplot(column='Total Revenue', by='Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Box Plot of Total Revenue by Product Category')
plt.suptitle('')
plt.xticks(rotation=45)
plt.show()
```

Violin Plot

```python

plt.figure(figsize=(12, 6))
sns.violinplot(x='Product Category', y='Total Revenue', data=df)
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.title('Violin Plot of Total Revenue by Product Category')
plt.xticks(rotation=45)
plt.show()
```
